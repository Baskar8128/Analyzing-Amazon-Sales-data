```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Amazon Sales data.csv')
data.head()
```

```python
data.describe()
```

```python
data.isnull().any()
```

```python
sns.countplot(data=data, x= 'Order Priority')
plt.title('SUM OF ORDER PRIORITY', fontsize = 15)
plt.show()
```

```python
data['Order Priority'].value_counts()
```

```python
counts = data['Sales Channel'].value_counts()
counts.plot(kind='pie', figsize = (7, 7), autopct='%0.1f%%')
plt.xlabel('Sales Channel Counts')
plt.title('NUM OF SALES CHANNEL')
plt.show()
```

```python
sns.countplot(data = data, y = 'Region')
plt.title('REGION WISE ORDERS')
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data.head()
```

```python
data['Item Type'].unique()
```

```python
data['Month'] = data['Order Date'].dt.month
data['Year']=data['Order Date'].dt.year
data.head()
```

```python
sns.countplot(data = data, x = 'Month')
plt.title('MONTH WISE ORDERS')
plt.xlabel('Month')
plt.ylabel('Num of Orders')
```

```python
sns.countplot(data = data, y = 'Item Type')
plt.title('ITEM TYPES ORDERS',fontsize=20)
plt.ylabel('Item Types',fontsize=15)
plt.xlabel('Num of Orders',fontsize=15)
```

```python
sns.countplot(data = data, x = 'Year')
plt.title('YEAR WISE ORDERS')
plt.ylabel('Num of Orders')
```

```python
df = data.groupby(data['Order Date'].dt.year)['Total Revenue'].sum()
df.plot(kind='pie', figsize=(7, 7), autopct='%0.1f%%')
plt.title('Total Revenue Year-wise', fontsize = 15)
plt.ylabel('Total Revenue', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(7, 7))
data.groupby(data['Order Date'].dt.month)['Total Revenue'].sum().plot(kind='pie',autopct='%0.1f%%')
plt.title('Total Revenue month-wise', fontsize = 15)
plt.ylabel('Total Revenue', fontsize = 15)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(15,8)
data.groupby([data['Order Date'].dt.year, data['Month']])['Total Revenue'].sum().unstack().plot(kind='bar')
plt.title('Total Revenue Year & Month-wise', fontsize = 15)
plt.xlabel('Year Month Wise', fontsize = 20)
plt.ylabel('Total Revenue', fontsize = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(15,8)
data.groupby(data['Item Type'])['Total Revenue'].sum().plot(kind='bar')
plt.title('Total Revenue Item-wise', fontsize = 15)
plt.xlabel('Item types', fontsize = 15)
plt.ylabel('Total Revenue', fontsize = 15)
plt.show()
```

```python
df = data.groupby(data['Item Type'])['Units Sold'].sum()
df.plot(kind='bar', figsize=(10,7))
plt.title('Units sold item type wise', fontsize = 15)
plt.ylabel('Units Sold', fontsize = 15)
plt.xlabel('Item Type', fontsize = 15)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(20,10)
df = data.groupby(data['Country'])['Total Profit'].sum()
df.plot(kind='bar')
plt.title('country wise total profit', fontsize = 15)
plt.ylabel('Total Profit', fontsize = 15)
plt.xlabel('Country', fontsize = 20)
plt.show()
```
